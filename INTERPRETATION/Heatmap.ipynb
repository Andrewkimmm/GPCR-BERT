```python
# Load the data
head = 0
atn_test = np.load(f"/content/drive/MyDrive/PRotBert/npxxy_head/final_NPxxY_weight_and_sequences_test(head{head}).npy", allow_pickle=True)
weight = atn_test[1]['attention']

# Set up plot parameters
width = 1.5
tick_width = 0.5
fig, ax1 = plt.subplots(figsize = [5,3], dpi = 600)


#for _, spine in ax1.spines.items():
#    spine.set_visible(True)
#    spine.set_linewidth(tick_width)

ax1.tick_params(direction = 'in', width = tick_width)
#plt.xticks(np.arange(0, 370, 100), np.arange(0,400,100), fontsize = 10)
#plt.yticks(np.arange(0, 370, 100), np.arange(0,400,100), fontsize = 10)


# Define a custom colormap
colors = ["white", "darkblue", "black"]  # replace the colors as needed
cmap = LinearSegmentedColormap.from_list("custom", colors, N=256)

# Plot heatmap
ax1 = sns.heatmap(weight, cmap=cmap)
ax1.set_xticks([100,200,300])
ax1.set_xticklabels(['100', '200', '300'], fontsize = 10, rotation = 'horizontal')
ax1.set_yticks([100,200,300])
ax1.set_yticklabels(['100', '200', '300'], fontsize = 10)

for _, spine in ax1.spines.items():
      spine.set_visible(True)
      spine.set_linewidth(width)


plt.show()
```
